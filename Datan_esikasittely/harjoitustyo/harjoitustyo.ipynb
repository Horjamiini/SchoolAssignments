## Vaihe 1

#### Tehtävä 1. Hae Digitrafficin REST-rajapintapalvelusta (JSON-formaatissa) aktiiviset junat pääteosoitteella `/live-trains` Helsingin asemalta, joiden hakuparametrit ovat seuraavat:
* arrived_trains=10
* departed_trains=10
* include_nonstopping=False

```python
import requests
import pandas as pd
import json
```

```python
arrived_trains="10"
departed_trains="10"
include_nonstopping="False"
url = "https://rata.digitraffic.fi/api/v1/live-trains/station/HKI?arrived_trains={0}&departed_trains={1}&include_nonstopping={2}".format(arrived_trains,departed_trains,include_nonstopping)
req = requests.get(url=url)
req.status_code

req_json = req.json()
req_json
```

#### Tehtävä 2. Tämän jälkeen lue kyseinen data Pandasin dataframeen ilman muokkaamista, jolloin sarakkeiden otsikot ovat seuraavanlaiset:

 `cancelled`,
 `commuterLineID`,
 `departureDate`,
 `operatorShortCode`,
 `operatorUICCode`,
 `runningCurrently`,
 `timeTableRows`,
 `timetableAcceptanceDate`,
 `timetableType`,
 `trainCategory`,
 `trainNumber`,
 `trainType`,
 `version` 

```python
df = pd.json_normalize(req_json)
df
```

#### Tehtävä 3. Poista kaikki muut sarakkeet paitsi `timeTableRows` ja `trainNumber`

```python
train_df = df[["timeTableRows","trainNumber"]]
train_df
```

#### Tehtävä 4. Tämän jälkeen pura `timeTableRows` sarakkeen arvot jokaisesta junasta erikseen (sarakkeen arvo jokaiselle junalle sisältää useampia JSON-objekteja) ja lisää dataan mukaan junan tiedot yksilöivä `trainNumber`

```python
timetable_df = pd.json_normalize(train_df['timeTableRows'].explode())
timetable_df
```

#### Tehtävä 5. Pudota sitten seuraavat sarakkeet datasta: `causes`, `estimateSource`, `liveEstimateTime`, `differenceInMinutes`, `commercialStop` ja `trainReady`

```python
timetable_df = timetable_df.drop(['causes','estimateSource','liveEstimateTime','differenceInMinutes','commercialStop','trainReady.source','trainReady.accepted','trainReady.timestamp'],axis=1)
timetable_df
```

#### Tehtävä 6. Täytä jokaisen rivin sarakkeen `actualTime` puuttuvat arvot `scheduledTime` -sarakkeen arvolla.

```python
timetable_df['actualTime'].fillna(timetable_df['scheduledTime'],inplace=True)
timetable_df
```

## Vaihe 2

#### Tehtävä 7. Pura tämän kansion sisällä olevan `data.html` -tiedoston `table`-elementin solujen arvot käyttäen regex-syntaksia ja yhdistä ne vaiheessa 1 luotuun dataframeen. Huom: tehtävässä tulee käyttää vain regexiä (Esim. Pandasin read_html()-metodin käyttö on kielletty)!

```python
import re

data = open("data.html","r").read()
rh = r'<th>(.*?)</th>'
rd = r'<td>\s*(.*?)\s*</td>'

datah = re.findall(rh,data)
datad = re.findall(rd,data)

search_list = []

for i in range(0,len(datad),10):
    search_list.append(datad[i: i +10])

html_df = pd.DataFrame(search_list,columns=datah)
html_df

complete_df = pd.concat([timetable_df,html_df],axis=0).reset_index(drop=True)
complete_df
```

## Vaihe 3

#### Tehtävä 8. Tee SQLite tietokanta ja lisää sinne vaiheen 2 tuloksena saatu dataframe.


```python
import sqlite3
database_connection = sqlite3.connect("trains.db")
C = database_connection.cursor()

complete_df.to_sql(name="traintimetable",con=database_connection,if_exists="fail",index=False)

trains_df = pd.read_sql_query('SELECT * FROM traintimetable',database_connection)
trains_df
```
