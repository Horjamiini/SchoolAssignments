```python
from itertools import combinations

def makecomb(l):
    n = len(l)
    combos = []
    for i in range(1, n+1):
        for combo in combinations(l, i):
            if sorted(combo) in combos:
                continue
            sub_combos = makecomb(list(set(l) - set(combo)))
            for sub_combo in sub_combos:
                new_combo = sorted(combo + sub_combo)
                if new_combo not in combos and len(new_combo) == n:
                    combos.append(new_combo)
    return combos
makecomb([1,2,3])
```
